```python
import pandas as pd 
```

```python
df=pd.read_csv('data\my_data.csv')
```

```python
df.head()
```

```python
df=df.drop(labels=['ID','Dt_Customer'],axis=1)
```

```python
X=df.drop(labels=['Response'],axis=1)
Y=df[['Response']]
```

```python
X.head()
```

```python
Y.head()
```

```python
#Define which columns should be ordinal -encoded and which should be scaled 
categorical_cols=X.select_dtypes(include='object').columns
numerical_cols=X.select_dtypes(exclude='object').columns
```

```python
education_categories=['Graduation', 'PhD', 'Master', 'Basic', '2n Cycle']
marital_Status_categories=['Single', 'Together', 'Married', 'Divorced', 'Widow', 'Alone','Absurd', 'YOLO']
       
```

```python
from sklearn.impute import SimpleImputer ## Handling Missing Values
from sklearn.preprocessing import StandardScaler # Handling Feature Scalling 
from sklearn.preprocessing import OrdinalEncoder # Ordinal Encoding

##Pipelines
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
##Numerical Pipelines
num_pipeline =Pipeline(
    steps=[
        ('imputer',SimpleImputer(strategy='median')),
        ('scaler',StandardScaler())

    ]

)
# Categorical Pipeline
cat_pipeline =Pipeline(
    steps=[
        ('imputer',SimpleImputer(strategy='most_frequent')),
        ('OrdinalEncoder',OrdinalEncoder(categories=[education_categories,marital_Status_categories])),
        ('scaler',StandardScaler())
    ]
)

preprocessor=ColumnTransformer([
    ('num_pipeline',num_pipeline,numerical_cols),
    ('cat_pipeline',cat_pipeline,categorical_cols)

])
```

```python
## Train Test Spilt

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.30,random_state=30)
```

```python
X_train=pd.DataFrame(preprocessor.fit_transform(X_train),columns=preprocessor.get_feature_names_out())
Y_test=pd.DataFrame(preprocessor.transform(X_test),columns=preprocessor.get_feature_names_out())
```
